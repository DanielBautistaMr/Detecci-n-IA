```python
# Esta línea de código cuenta las ocurrencias de valores únicos en la columna 'BARRIOS_HECHO' de un DataFrame 'df' y
# devuelve un objeto que muestra cuántas veces aparece cada valor en esa columna.
# Es útil para obtener una idea de la distribución de los datos en esa columna.
df.BARRIOS_HECHO.value_counts()
```

```python
# Esta línea de código realiza una fusión (merge) entre dos DataFrames, 'df' y 'dfbarrios', utilizando la columna 'NOM_COM'
# como clave de fusión. Esto combinará las filas de ambos DataFrames en función de los valores coincidentes en la columna 'NOM_COM'.
df = pd.merge(df, dfbarrios, on="NOM_COM")

df
```

```python
df.columns
```

```python
# Esta línea de código elimina las columnas 'Unnamed: 0' y 'loc' del DataFrame 'df'.
# El argumento 'axis=1' especifica que las columnas deben eliminarse en lugar de las filas.
# El argumento 'inplace=True' indica que la operación debe realizarse directamente en el DataFrame 'df',
# lo que significa que el DataFrame 'df' se modificará y no se creará uno nuevo.
df.drop(['Unnamed: 0', 'loc'], axis=1, inplace=True)
```

```python
# Esta línea de código crea una nueva columna llamada 'FECHA_COMPLETA' en el DataFrame 'df'.
# Combina el contenido de las columnas 'FECHA_HECHO' y 'HORA_HECHO' y los concatena en un único valor de cadena
# separado por un espacio en blanco, creando así una nueva columna que contiene la fecha y la hora en un formato completo.
df['FECHA_COMPLETA'] = df["FECHA_HECHO"] + ' ' + df["HORA_HECHO"]

df
```

```python
#convertir fecha completa en formator calendario

df=df.astype({"FECHA_HECHO":"datetime64[ns]"})
```

```python
df.info()
```

```python
# Esta línea de código convierte la columna 'FECHA_HECHO' del DataFrame 'df' en un tipo de datos datetime.
# Utiliza el método 'pd.to_datetime()' para realizar la conversión.
# El argumento 'format="DD/MM/YYYY"' se utiliza para especificar el formato de la fecha original en la columna.
# Asegúrate de que el formato especificado coincida con el formato real de las fechas en la columna.
df["FECHA_HECHO"] = pd.to_datetime(df["FECHA_HECHO"], format="DD/MM/YYYY")
```

```python
#3 cuantos delitos por año se tiene

cantidadaño=df.groupby(df["FECHA_HECHO"].dt.year)["DESCRIPCION_CONDUCTA"].count().to_frame()
cantidadaño
```

```python
df_localizacion=df.localizacion.str.split(',',expand=True)
df_localizacion=df_localizacion.rename(columns={0:'lat',1:'lon'})
df_localizacion=df_localizacion.replace('\[','',regex=True).replace(']','',regex=True).astype(float)
df_localizacion.dtypes
```

```python
df.drop(['localizacion','LOCALIDAD'],axis=1,inplace=True)
```

```python
#agrupar para hacer mapa de calor de delitos x comuna
cantidadComuna=df.groupby(['NOM_COM','lat','lon'])['DESCRIPCION_CONDUCTA'].count().to_frame()
cantidadComuna['NOM_COM']=cantidadComuna.index.get_level_values(0)
cantidadComuna['lat']=cantidadComuna.index.get_level_values(1)
cantidadComuna['lon']=cantidadComuna.index.get_level_values(2)
cantidadComuna=cantidadComuna.reset_index(drop=True)
cantidadComuna
```

```python
fig = px.density_mapbox(cantidadComuna, lat = 'lat', lon = 'lon',z='DESCRIPCION_CONDUCTA',
                        radius = 50,
                        hover_name='NOM_COM',
                        color_continuous_scale='rainbow',
                        center = dict(lat = 7.12539, lon = -73.1198),
                        zoom = 12,
                        mapbox_style = 'open-street-map')
fig.show()
```
